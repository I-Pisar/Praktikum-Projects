# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Вам нужно научиться определять рыночную стоимость объектов недвижимости. Для этого проведите исследовательский анализ данных и установите параметры, влияющие на цену объектов. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта и других объектов — эти данные автоматически получены из геосервисов. Количество парков и водоёмов также заполняется без участия пользователя.

**Цель исследования:**
1. Изучить рыночную стоимость объектов недвижимости Санкт-Петербурга и Ленобласти и динамику стоимости за исследуемый период (2014-2019 гг.).
2. Выявить аномалии и возможную мошенническую деятельность при помощи автоматизированных систем.
3. Определить главные показатели, влияющие на ценообразование недвижимости в Санкт-Петербурге и Ленобласти.

**Ход исследования:**

Сначала необходимо прочитать данные из датасета *'/datasets/real_estate_data.csv'* и оценить их качество и читаемость. Затем нужно будет произвести этап предобработки для первичной оценки ошибок и аномалий, по возможности исправляя их и сохраняя максимальное количество данных для изучения. Затем я добавлю новые показатели в датасет, необходимые и полезные для проведения исследовательского анализа данных. Также на этом этапе нужно будет произвести валидацию данных для достижения решения с необходимым качеством итогового результата.

Мое исследование пройдет в шесть этапов:
 - Обзор данных.
 - Предобработка данных.
 - Расчеты и добавление новых показателей.
 - Исследовательский анализ данных.
 - Изучение главных факторов, влиящюих на ценообразование недвижимости.

## Предобработка данных

Для начала приведем названия столбцов к общему красивому виду.

```python
data = data.rename(columns = {'cityCenters_nearest':'city_centers_nearest',
                       'parks_around3000':'parks_around_3000',
                       'ponds_around3000':'ponds_around_3000'})
```

```python
na = data.isna().sum()
na_perc = round((data.isna().mean()*100),2)

display(pd.DataFrame({'Кол-во пропусков': na,'% пропусков': na_perc}).style.background_gradient('coolwarm').format('{:.2f}'))
        
print("Всего колонок: {}, Без пропусков: {},с пропусками: {}".format(len(na), (na==0).sum(), (na!=0).sum() ))
#выведем количество пропущенных значений для каждого столбца датафрейма
```

Больше всего пропусков в **is_apartment**, меньше всего - в **locality_name**. Нам предстоит по порядку обработать пропущенные значения, заполнив их, где это возможно.

```python
sorted(data['ceiling_height'].unique()) #выведем уникальные значения столбца с данными о 
                                        #высоте потолка в квартирах, отсортированные по возрастанию
```

Минимальная высота потолков в многоквартирных домах регламентируется **СНиП 31–01–2003** Российской Федерации и составляет 2.5 метра. Соответственно, все значения меньше можно удалить, смело окрестив их аномальными.

Максимальная высота стен не регламентирована, однако новые дома строятся с показателем не более 3,2 м. Исключения составляют квартиры элитных жилых комплексов, в которых потолки могут достигать высоты до 4 м. Поэтому примем высоту в 4 метра за максимально возможное значение.

Значения 25.0, 26.0, 27.0, 27.5, 32.0 выглядят как ошибочные, неправильно введенные. Их можно исправить, разделив все значения на 10.



```python
data['ceiling_height'] = data['ceiling_height'].apply(lambda x: x/10 if x>20 else x)
#применим трансформацию ко всем значениям столбца, которые превышают значение 20
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
#заменим пропущенные значения на медианные

data = data.query('ceiling_height >= 2.5 and ceiling_height <= 4.0')
#отфильтруем таблицу по значениям высоты потолка между 2.5 и 4 метра
```

Теперь снова выведем список уникальных значений столбца, чтобы проверить проведенную обработку данных.

```python
sorted(data['ceiling_height'].unique())
```

Пропущенные значения заменены и все значения отфильтрованы, можно приступать к следующему столбцу - **floors_total**.

Количество пропущенных значений в данном столбце крайне мало по отношению к общему количеству строк, также этот параметр не зависит от других и замена значений не представляется возможной. Можно смело удалять строки с пропущенными данными.

```python
data = data.dropna(subset=['floors_total']) #применяем метод dropna() к столбцу со значением количества этажей
```

Также нужно поменять тип данных на целочисленный, так как количество этажей выражается в целых числах.

```python
data['floors_total'] = data['floors_total'].astype('int') #меняем тип данных на целочисленный

data['floors_total'].isna().sum() #проверяем отсутствие пропущенных значений
```

Далее идет столбец с данными о жилой площади - **living_area**. Этот параметр сильнее всего зависит от общей площади, поэтому заменим пропущенные значения на отношение средней жилой площади к средней общей жилой площади квартиры. В столбце с данными об общей площади нет пропусков.

```python
data['living_area'].isna().sum() #выведем количество пропущенных значений в столбце с жилой площадью квартир
```

```python
data['living_area'].fillna((data['living_area'].mean() / data['total_area'].mean() *
                                                 data['total_area']), inplace = True)
#используем цикл, который заменит пропуски на отношение средней жилой площади 
#к средней общей площади квартиры, помноженной на общую площадь
```

```python
data['living_area'].isna().sum() #проверяем количество пропусков после обработки
```

Теперь обработаем данные в **is_apartment**. В данном столбце тип данных - *object*, хотя этот параметр подразумевает булев тип данных - True или False. Заменим тип данных на корректный.

```python
data['is_apartment'] = data['is_apartment'].astype('bool')
```

Далее заменим пропуски. Можно смело предположить, что большинство пропущенных значений означают, что тип квартиры не является апартаментами, поэтому заменим их на False.

```python
data['is_apartment'] = data['is_apartment'].fillna(value=False) #заменим пропуски на False

data['is_apartment'].isna().sum() #проверим отсутствие пропусков
```

На очереди обработка пропусков в столбце **kitchen_area**. Заменим их значением отношения средней площади кухни к средней общей площади квартиры, помноженной на общую площадь.

```python
data['kitchen_area'].isna().sum() #выведем колчество пропусков
```

```python
data['kitchen_area'].fillna((data['kitchen_area'].mean() / data['total_area'].mean() *
                                                 data['total_area']), inplace = True)
#заменим пропуски с аргументом inplace=True для замены оригинального датафрейма
```

```python
data['kitchen_area'].isna().sum() #проверим отсутствие пропусков
```

Далее обработаем пропуски в **balcony**. Количество балконов не зависит от прочих факторов, поэтому пропуски в данных скорее всего означают отсутствие балконов.

```python
data['balcony'].isna().sum() #выведем количество пропусков
```

```python
data['balcony'] = data['balcony'].fillna(0) #заменяем пропуски на ноль

data['balcony'].isna().sum() #и проверяем отсутствие пропусков
```

```python
data['balcony'] = data['balcony'].astype('int')
```

В **locality_name** ничтожное количество пропусков по отношению ко всему массиву данных, к тому же сложно придумать логичную замену, поэтому удалим строки с пропусками.

```python
data['locality_name'].isna().sum()
```

```python
data = data.dropna(subset = ['locality_name'])

data['locality_name'].isna().sum()
```

Теперь рассмотрим **airports_nearest**. Пропущенные значения могли появиться из-за ошибки системы, рассчитывающей расстояние до ближайшего аэропорта, например, если расстояние превышает определенную величину.

```python
display(data['airports_nearest'].describe()) #выведем данные о значениях столбца airports_nearest

data['airports_nearest'].isna().sum() #посмотрим количество пропусков
```

В столбце с удаленностью от аэропорта присутствуют нулевые значения, датафрейм стоит очистить от таких значений. Количество пропусков снизилось вследствие фильтрации значений других показателей. Логичную замену этим данным сложно представить. Поэтому я считаю, что пропуски следует оставить, так как они все еще составляют значительную часть данных.

```python
data = data.query('airports_nearest != 0') #очистим таблицу от строк с нулевым значением удаленности от аэропорта
```

```python
data['airports_nearest'].describe() #посмотрим обновленные значения
```

Нулевое значение было всего лишь в одной строке. Можно переходить к следующему показателю - **city_centers_nearest**. Пропущенные данные могли появиться из-за ошибки в системе, которая рассчитывала расстояние в определенном радиусе.

```python
display(data['city_centers_nearest'].describe()) 

data['city_centers_nearest'].isna().sum()
```

Логичную замену для пропусков в данном показателе сложно представить, пока что оставим их. Переходим к **parks_around_3000**. 
Пропуски в данном столбце могли появиться потому, что в данном радиусе попросту нет парков. Заменим пропуски на 0. То же самое и с прудами **ponds_around_3000**. Также нужно заменить в них тип данных на целочисленный, т.к. количество парков и прудов выражается в целых числах.

```python
data['parks_around_3000'] = data['parks_around_3000'].fillna(0).astype('int')

data['ponds_around_3000'] = data['ponds_around_3000'].fillna(0).astype('int')
```

```python
display(data['parks_around_3000'].isna().sum())

data['ponds_around_3000'].isna().sum()
```

В столбцах **parks_nearest** и **ponds_nearest** пропущенные данные скорее всего вызваны тем, что в них попросту не были заполнены данные. Не будем трогать их, тем более что данные показатели нам не нужны для расчета итоговых зависимостей стоимости.


**days_exposition**: пропуски в данном столбце могут быть вызваны некорректной работой ресурса, например квартира была продана, но это было не зафиксировано программой. По этой же причине могут проглядываться аномально большие значения. Хоть этот показатель и очень важен, но он не зависит от других и поэтому не представляется возможным придумать логичную замену. Оставим пропуски как есть.

```python
data['days_exposition'].max()
#посмотрим на максимальное значение показателя
```

```python
data['days_exposition'] = data['days_exposition'].astype('int64', errors='ignore') #меняем тип данных на int

data['days_exposition'].describe()
```

```python
data['days_exposition'].isna().sum() #проверим количество пропусков
```

Теперь поменяем тип данных в **first_day_exposition**. Нужно поменять его на *datetime*, так как данный столбец содержит данные о дате, в которую было выставлено объявление о продаже квартиры.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Проверим количество комнат в квартирах-студиях. Построим сводную таблицу с индексом по количеству комнат, группировкой по значениям True и False для показателя **studio**, в values можно использовать любой показатель, где количество строк соответствует полному количеству строк датафрейма. Если количество комнат не равно 1, значит в данных ошибка и нужно поменять значение на 1, т.к. квартиры-студии всегда имеют 1 комнату.

```python
pd.pivot_table(data, index = 'rooms', columns = 'studio', values = 'total_area', aggfunc = 'count')
```

```python
data.loc[data['studio'] == True, 'rooms'] = 1 #заменяем значение комнат на 1 для всех строк, где показатель studio = True.
```

В сводной таблице были замечены другие проблемы, требующие обработки. Есть квартиры с нулевым показателем количества комнат, что невозможно. Также после 9 комнат значения выглядят неправдоподобно и не представляют статистической ценности для анализа. Примем 9 комнат за максимальное число.

```python
data = data.query('rooms > 0 and rooms <= 9') #отфильтруем датафрейм по количеству комнат между 1 и 9
```

```python
pd.pivot_table(data, index = 'rooms', columns = 'studio', values = 'total_area', aggfunc = 'count') 
#снова выведем сводную таблицу для проверки значений
```

Данные выглядят корректно и правдоподобно. Теперь переходим к названиям населенных пунктов в столбце **locality_name**. Выведем список уникальных значений и посмотрим наличие неявных дубликатов в категориальных переменных.

```python
sorted(data['locality_name'].unique())
```

### Факторы, которые больше всего влияют на общую (полную) стоимость объекта.

Изучим, как сильно влияют на общую стоимость объекта следующие параметры:
- общая площадь квартиры
- жилая площадь
- площадь кухни
- количество комнат
- тип этажа, на котором расположена квартира
- даты размещения объявления (день, месяц, год).

Для этого нам необходимо рассчитать коэффициент корреляции Пирсона для этих факторов, а также построить графики для наглядности и подтверждения либо опровержения гипотез.

```python
data[['price_mil','total_area','living_area', 'kitchen_area',
     'rooms', 'weekday_exposition', 'month_exposition','year_exposition']].corr()
#выведем таблицу с коэффициентами корреляции для всех параметров, кроме типа этажа, тк это нечисловой параметр
```

По таблице с коэффициентами корреляции видно, что цена квартиры сильнее всего взаимосвязана с общей и жилой площадью. Связь сохраняется и для площади кухни, количества комнат, но она выражена слабее.

Рассчитывать коэффициент корреляции для даты размещения о продаже было не самым корректным решением, так как цена зависит от них не линейно. Зависимость цены от этого фактора рассмотрим далее при помощи графиков.

Перед постройкой графиков напишем функции для упрощения дальнейших расчетов.

```python
#напишем функцию для рассчёта коэффициента корреляции с округлением до 3 знаков после запятой
def show_corr(column):
    return data[column].corr(data['price_mil']).round(3)

#напишем функцию, которая будет создавать сводные таблицы в зависимости 
#от заданного параметра по отношению к медианной стоимости квартиры
def pivot_med(column):
    return data.pivot_table(index = column, values = 'price_mil', aggfunc = 'median')
```

**Зависимость стоимости квартиры от общей площади**

```python
print('Коэффициент корреляции между стоимостью квартиры и общей площадью:', show_corr('total_area'))

pivot_med('total_area').plot(
    style='o',
    grid=True,
    figsize=(6,6),
);

plt.title('Зависимость стоимости от общей площади квартиры')
plt.ylabel('Цена в млн')
plt.xlabel('Общая площадь')
plt.show()
```

Прослеживается сильная линейная зависимость стоимости объекта от его общей площади. Это вполне логично, так как общая площадь квартиры - один из приоритетных факторов, на которые обращается внимание при покупке недвижимости.

**Зависимость стоимости квартиры от жилой площади**

```python
print('Коэффициент корреляции между стоимостью квартиры и жилой площадью:', show_corr('living_area'))

pivot_med('living_area').plot(
    style='o',
    grid=True,
    figsize=(6,6),
);

plt.title('Зависимость стоимости от жилой площади квартиры')
plt.ylabel('Цена в млн')
plt.xlabel('Жилая площадь')
plt.show()
```

Зависимость сохраняется, но она выражена чуть менее заметно. Чем больше жилая площадь - тем больше цена.

**Зависимость стоимости квартиры от площади кухни**

```python
print('Коэффициент корреляции между стоимостью квартиры и площадью кухни:', show_corr('kitchen_area'))

pivot_med('kitchen_area').plot(
    title='Зависимость стоимости от площади кухни', 
    style='o',
    grid=True,
    figsize=(6,6),
);

plt.title('Зависимость стоимости от площади кухни')
plt.ylabel('Цена в млн')
plt.xlabel('Площадь кухни')
plt.show()
```

Зависимость слабее, чем у жилой площади, но все еще сохраняется и она больше значения 0,5. Получается, что площадь кухни не играет решающей роли в ценообразовании, но все равно имеет свой вес.

**Зависимость стоимости квартиры от количества комнат**

```python
print('Коэффициент корреляции между стоимостью квартиры и количеством комнат:', show_corr('rooms'))

pivot_med('rooms').plot(
    grid=True,
    figsize=(6,6),
);

plt.title('Зависимость стоимости от количества комнат')
plt.ylabel('Цена в млн')
plt.xlabel('Количество комнат')
plt.show()
```

Зависимость стоимости от количества комнат выражена слабее, чем остальные факторы. В целом, зависимость сохраняется, но значение коэффициента корреляции Пирсона в 0,489 говорит о том, что это далеко не самый ценообразующий параметр. Это можно объяснить тем, что различные планировки позволяют создавать разное количество комнат при одинаковой площади квартиры, и прежде всего именно площадь влияет на стоимость объекта. Но график четко показывает нам линейную зависимость: чем больше комнат - тем дороже квартира!

**Зависимость стоимости квартиры от типа этажа, на котором находится квартира**

```python
pivot_med('floor_type').plot(
    title='Зависимость стоимости от типа этажа квартиры', 
    grid=True,
    figsize=(6,6),
);

plt.title('Зависимость стоимости от типа этажа квартиры')
plt.ylabel('Цена в млн')
plt.xlabel('Тип этажа')
plt.show()
```

Самые дешевые квартиры - те, которые располагаются на первом этаже. Это можно объяснить несколькими причинами. Одна из них - вопрос безопасности: ворам проще пробраться в квартиры на первом этаже, окна этих квартир чаще могут повреждаться хулиганами и тд. Другая причина - шум. Через первый этаж проходит наибольшее количество людей и из этих квартир лучше слышно, что происходит во дворе. Еще одна причина - страх быть затопленными соседями сверху, так как вода может попасть на 1 этаж с нескольких этажей сверху, а дальше ей уже некуда уходить, разве что в подвальное помещение, если оно имеется в доме.

Квартиры на последнем этаже хоть и дороже тех, что на первом, но тоже пользуются меньшим спросом, чем остальные. Это можно объяснить следующими факторами. Фактор шума: на последнем этаже располагается лифтовое оборудование, которое может издавать заметный уровень шума. Погодный фактор: квартиры на последнем этаже сильнее других подвержены ветрам и осадкам. Также присутствует страх протечек из-за некачественно выполненной или обработанной кровли. Также на последний этаж дольше добираться и перевозить мебель и бытовую технику. Особенно в случае поломки лифтового оборудования.
